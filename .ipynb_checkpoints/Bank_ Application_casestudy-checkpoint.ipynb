### Calculating the percentage of missing values for each column

```python
null_percen = app_data_df.isnull().sum() / len(app_data_df) * 100

# Calculating the count of missing values for each column
null_count = app_data_df.isnull().sum()

# Creating a DataFrame to store the column names, missing values count, and their percentage
missing_values_df = pd.DataFrame({
    'column_name': app_data_df.columns,
    'null Count': null_count,
    'null Percentage': null_percen
})
# Resetting the index and sorting the DataFrame by percentage of missing values in descending order
missing_values_df.reset_index(drop = True, inplace=True)
missing_values_df.sort_values(by = 'null Percentage', ascending = False)
```

### Identifying columns with more than 45% missing values

```python
columns_to_be_deleted = missing_values_df[missing_values_df['null Percentage'] > 45].column_name.to_list()
```

### Displaying the list of columns that have more than 45% missing values and will be considered for deletion

```python
columns_to_be_deleted
```

### Dropping columns with more than 45% missing values from the dataset

```python
app_data_df.drop(columns = columns_to_be_deleted, inplace = True)
```

# Handling of outlier in AMT_INCOME_TOTAL 
#### Filter the DataFrame to find entries where AMT_INCOME_TOTAL equals 117,000,000

```python
filtered_Total_income = app_data_df[app_data_df['AMT_INCOME_TOTAL'] == 117000000.00]
filtered_Total_income
```

### Step 1: Correct the Total Income Amounts
###  Filter the DataFrame for rows where occupation type is 'laborers' and the total income is incorrect

```python
app_data_df.groupby('OCCUPATION_TYPE')['AMT_INCOME_TOTAL'].sum()
```

### Calculate the adjusted average of total income for laborers

```python
avg_of_laborers = (9180604030.64-117000000.00)/(55186-1)
avg_of_laborers
```

### Replace the outlier value (117,000,000) with the calculated mean value (164,240.36)
### Use the mean value calculated previously as the replacement value

```python
app_data_df['AMT_INCOME_TOTAL'] = app_data_df['AMT_INCOME_TOTAL'].replace(117000000.00 ,164240.3557242004)

# Calculate the new mean of 'AMT_INCOME_TOTAL' after the replacement
app_data_df['AMT_INCOME_TOTAL'].mean()
```
